```python
import networkx as nx
```

```python
def answer_one():
    return nx.read_edgelist('email_network.txt', data=(('time',int),),create_using=nx.MultiDiGraph())
```

```python
def answer_two():
    G = answer_one()
    return len(G.nodes()), len(G.edges())
answer_two()
```

```python
def answer_three():    
    G = answer_one()
    return nx.is_strongly_connected(G), nx.is_connected(G.to_undirected())
answer_three()
```

```python
nx.draw_networkx(G)
```

```python
def answer_four():
    G = answer_one()
    wccs = nx.weakly_connected_components(G)
    return len(max(wccs, key=len))
answer_four()
```

```python
def answer_five():
    G = answer_one()
    sccs = nx.strongly_connected_components(G)
    return len(max(sccs, key=len))
answer_five()
```

```python
def answer_six():
    G = answer_one()
    scc_subs = nx.strongly_connected_component_subgraphs(G)
    G_sc = max(scc_subs, key=len)
    return G_sc
answer_six()
```

```python
def answer_seven():
    G = answer_six()
    return nx.average_shortest_path_length(G)
answer_seven()
```

```python
def answer_eight():
    G = answer_six()
    return nx.diameter(G)
answer_eight()
```

```python
def answer_nine():
    G = answer_six()
    return set(nx.periphery(G))
answer_nine()
```

```python
def answer_ten():
    G = answer_six()
    return set(nx.center(G))
answer_ten()
```

```python
def answer_eleven():
    G = answer_six()
    d = nx.diameter(G)
    peripheries = nx.periphery(G)
    max_count = -1
    result_node = None
    for node in peripheries:
        sp = nx.shortest_path_length(G, node)
        count = list(sp.values()).count(d)
        if count > max_count:
            result_node = node
            max_count = count
    return result_node, max_count
answer_eleven()
```

```python
def answer_twelve():
    G = answer_six()
    center = nx.center(G)[0]
    node = answer_eleven()[0]
    return len(nx.minimum_node_cut(G, center, node))
answer_twelve()
```

```python
def answer_thirteen():
    G = answer_six()
    undir_subgraph = G.to_undirected()
    G_un = nx.Graph(undir_subgraph)
    return G_un
answer_thirteen()
```

```python
def answer_fourteen():
    G = answer_thirteen()
    return nx.transitivity(G), nx.average_clustering(G)
answer_fourteen()
```
